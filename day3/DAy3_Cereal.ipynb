```python
cat_names = ["mfr", "type"]
dep_var = ["rating"]
features = list(set(df.columns)-{"name",dep_var[0]})
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[features], df[dep_var], test_size=0.2, random_state=42)
```

```python
X_train, y_train = X_train.values, y_train.values.ravel()
X_test, y_test = X_test.values, y_test.values.ravel()
```

```python
from catboost import CatBoostRegressor
```

```python
reg = CatBoostRegressor(iterations=200)
```

```python
reg.fit(X_train, y_train, cat_features=cat_names, eval_set=(X_test, y_test))
```
